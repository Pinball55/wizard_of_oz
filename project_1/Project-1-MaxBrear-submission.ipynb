### Challenge 1.1: Largest Palindrome
A palindromic number reads the same both ways. For example, 1234321 is a palindrome. The largest palindrome made from the product of two two-digit numbers is 9009 = 91 × 99. Find the largest palindrome made from the product of two three-digit numbers.

```python
#-----------Code-----------#

min_num = 1  # Inclusive Number - Lower end of the numbers in the range to test
max_num = 999  # Inclusive Number - Higher end of the numbers in the range to test
max_palindrome = 0

for n in range(min_num, max_num + 1):  # Cycle through the range of numbers from min to max
    for i in range(min_num, max_num + 1):  # For each n, cycle through a 2nd time to check multiples on n * i
        
        palindrome = i * n  # Starting palindrome; This is a variable that changes on the calc outcome 
        palindrome_string = str(palindrome)  # Create a string from the palindrome integer so a check can be performed to identify true palindromes
        
        if palindrome_string == palindrome_string[::-1]:  # Check if it's a palindrome by seeing if the number is the same forwards vs backwards
            if int(palindrome) > int(max_palindrome):  # Only updates the palindrome if the value is larger than the previous maximum
                max_palindrome = palindrome_string  # If it is a palindrome set the value to maximum_palindrome because we're looking for the biggest
                
        else:
            i = i  # If it's not a palindrome move onto the next number i
        
print("Maximum Palindrome = " + str(max_palindrome))  # Once all the multiples of n * i have been checked print the maximim palindrome found

#----My General Comments----#

# There must be a better way to essentially say 'do nothing and move on' rather than i = i
# Would it be bettwe to use a function to calculate the palidrome?
# To make it more efficient it wuld be better to start from the biggest number and work backwards, then stop at the first palindrome
```


### Challenge 1.2: Summation of Primes
The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17. Find the sum of all the primes below 2,000.

```python
#-----------Code-----------#

first_num = 2  # Inclusive Number - First in the series to sum
last_num = 1999 # Inclusive Number - Last in the series to sum
prime_total = 2  # Start the prime sum total at 2 so we include the first prime number 2 in the final putput

#print("2 Prime")  # Code can be added if you want to print the full list of primes

for n in range(first_num, last_num + 1):  # Cycle through the range to be reviewed
    for i in range(2, n):  # A 2nd range is needed to check if the number is a prime i.e. this is all the numbers between 2 and one less than the number being tested to see if it's a prime
        if n % i == 0:  # If the number n divided by any number 1 divides evenly with no remainder then it's not a prime so move to the next number
            #print(str(n) + " Not Prime")  # Code can be added if you want to print the full list of primes
            break  # break loop and move to the next value of n
        elif n == i + 1:  # If n / i does give a remainder it could be a prime so keep checking the rest of the i values. Once you get to n == i + 1 you've tried all the combinations so it's confirmed as a prime number
            #print(str(n) + " Prime")  # Code can be added if you want to print the full list of primes
            prime_total = prime_total + n  # add the prime number to the total
            break  # break loop and move to the next value of n
           
print(str(prime_total) + " Sum of Primes")   # print the final total

#----My General Comments----#

# I could adjust the code at some point so the first prime 2 is included in the calculation rather than added separately
```

### Challenge 1.3: Multiples of 3 and 5
If we list all of the natural numbers below 10 that are multiples of 3 or 5, we get 3, 5, 6, and 9. The sum of these multiples is 23. Find the sum of all the multiples of 3 and 5 below 1,000.

```python
#-----------Code-----------#

first_num = 0  # Inclusive Number - First in the series to check
last_num = 999  # Inclusive Number - Last in the series to check
num_total = 0  # the final total, starting at 0 when the code initially runs
#numbers_list = []  # Not being used but could be introduced in a new version that uses lists

for n in range(first_num, last_num + 1):  # Loop through the range required
    
    if n % 3 == 0:  # If the value is a multiple of 3 i.e. no remainder
        #print(n)  # Code can be added if you want to print the values of n
        #numbers_list.append(n)  # Not being used but could be introduced in a new version that uses lists
        num_total = num_total + n  # add the number to the total
    
    elif n % 5 == 0:  # Else if the value is a multiple of 5 i.e. no remainder
        #print(n)  # Code can be added if you want to print the values of n
        #numbers_list.append(n)  # Not being used but could be introduced in a new version that uses lists
        num_total = num_total + n  # add the number to the total

print(num_total)  # Print the final total

#-----------Alternate Solution-----------#

first_num = 0
last_num = 999
num_total = 0

for n in range(first_num, last_num + 1):
    
    if n % 3 == 0 or n % 5 == 0:
        num_total += n 

print(num_total)

#----My General Comments----#

# An alternative solution might be to create a list of numbers, appending each value and then multiplying them at the end
```

### Challenge 2: FizzBuzz
Write a program that prints all of the numbers from 1 to 100. For multiples of 3, instead of the number, print "Fizz;" for multiples of 5, print "Buzz." For numbers that are multiples of both 3 and 5, print "FizzBuzz." Bonus: If that's too easy, create three unique solutions. Sometimes you'll encounter unexpected constraints, so make sure you're able to think creatively!

```python
#-----------Code-----------#

first_num = 1  # Inclusive Number - First in the series to check
last_num = 100  # Inclusive Number - Last in the series to check

#---Solution 1---#

print("----Solution 1----")

for n in range(first_num, last_num + 1):  # Cycle through the range
    
    if n % 3 == 0 and n % 5 == 0:  # Start by checking if the number is a multiple of both 3 and 5
        print("FizzBuzz")  # If true then print FizzBuzz
            
    elif n % 3 == 0:  # Then check if it's just a multiple of 3
        print("Fizz")  # If true then print Fizz
    
    elif n % 5 == 0:  # Finally check if it's just a multiple of 5
        print("Buzz")  # If true then print Buzz
        
    else:  # If it's none of the above multiple then just print the number
        print(n)
        
print("")
        
#---Solution 2---#

print("----Solution 2----")

def fizzbuzzchecker(n):  # Define a Function to perform the calculation first
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")   
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:  
        print(n)     

num_list = list(range(first_num, last_num + 1))  # Use a list instead of a simple range.

for n in num_list:  # Execute the function
    fizzbuzzchecker(n)
    
print("")
        
#---Solution 3---#

print("----Solution 3----")

# Solutions uses the method a * a string (Fizz, Buzz or both) by 0 to remove it from the printed list

print("\n".join(["Fizz"*(number % 3 == 0) + "Buzz"*(number % 5 == 0) or str(number) for number in range(first_num, last_num + 1)]))
```

### Challenge 3: String Compressor
Implement a method to perform basic string compression using the counts of repeated characters. (This is called run-length encoding.) For example, the string "aabcccccaaa" would become a2b1c5a3. If the “compressed” string would not become smaller than the original string, your method should return the original string. You can assume the string has only uppercase and lowercase letters (a–z). Specify whether your solution is case sensitive or case insensitive and what you would need to change to make it the other.

```python
#-----------Code-----------#

# This code is quiet inefficient - With more time I could definately refine it but it still works :-)

counter = 1  # Counter starts at 1 count
input_string = "aabcccccaaa"  # The string to be compressed
comp_string = ""  # Compressed string starts blank 

for i in range(len(input_string) - 1):  #  Count the number of char in string and loop through these
    if i == (len(input_string) - 2):  # First if loop required to deal with last two characters of the string as they are a special case as the last character can't be compared with the last + 1 and therefore breaks the code
        if input_string[i] == input_string[i + 1]:  # If the last two characters are the same add + 1 to counter and then add to string
            counter = counter + 1
            comp_string = comp_string + input_string[i] + str(counter)
            counter = 1         
        else:            
            comp_string = comp_string + input_string[i] + str(counter) + input_string[i + 1] + str("1")  # if the last two chars are not the same you need to also add the final letter with a '1'
            counter = 1           
    elif input_string[i] == input_string[i + 1]:  # For all other characters in the letter next to the current letter is the same, add to the counter
        counter = counter + 1         
    else:
        comp_string = comp_string + input_string[i] + str(counter)  # If the next letter is different, add the letter and count to the final string
        counter = 1
        
if len(comp_string) > len(input_string):  # if the new string is longer than the original print the original
    print(input_string)
else: print(comp_string)
        
#----My General Comments----#

# Solution is case sensitive
# To remove the case sensitivity you would first convert the string to lower case using the str.lower() function
```
